```python
#machine learning algorithm is categorized into 3 main categories
# 1-supervised 2- unsupervised 3-reinforcement
# supervised learning is where the model is trained on labeled data
# unsupervised learning is where the model is trained on unlabeled data
# reinforcement learning is where the model learns by interacting with the environment

#K means clustering algorithm is used for the unsupervised data
# 1- start with k centroids by putting them in at some random place let k=2
# 2-compute the distance of each point from the centroid and cluster them accordingly
# adjust centroid so that they become center of gravity for given cluster
# 4- again recluster every point based on their distance with centroid
# 5- again adjust the centroids 
# 6- recompute clusters and repeat this till data points stop changing the clusters

# how to identify the corerct number of clusters (k) ? algo method
 
# SSE = sum of square errors you calculate for each cluster
# then plot sse vs k graph
# then find the elbow point of the graph which will give u the good cluster number

```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
df= pd.DataFrame({
    'Name':['Rob','Michael','mohan','ismail','kory','gautam','david','andrea','brad','angelina','donald','tom','arnold','jared','stark','ranbir','dipika','priyanka','nick','alia','sid','abdul'],
    'age':[27,29,29,28,42,39,41,38,36,35,37,26,27,28,29,32,40,41,43,39,41,39],
    'income':[70000,90000,61000,60000,150000,155000,160000,162000,156000,130000,137000,45000,48000,51000,49500,53000,65000,63000,64000,80000,82000,58000]

})
df
```

```python
plt.scatter(df['age'],df['income'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['age','income']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]

plt.scatter(df1['age'],df1['income'],color='green')
plt.scatter(df2['age'],df2['income'],color='red')
plt.scatter(df3['age'],df3['income'],color='blue')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Cluster 1','Cluster 2','Cluster 3'])
```

```python
scaler= MinMaxScaler()
scaler.fit(df[['income']])
df['income']=scaler.transform(df[['income']])
scaler.fit(df[['age']])
df['age']=scaler.transform(df[['age']])
```

```python
df
```

```python
km = KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['age','income']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]

plt.scatter(df1['age'],df1['income'],color='green')
plt.scatter(df2['age'],df2['income'],color='red')
plt.scatter(df3['age'],df3['income'],color='black')
km.cluster_centers_ #give the centroids of the clusters
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',s=200)
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(['Cluster 1','Cluster 2','Cluster 3'])
```

```python
# sum of square error calculation
k_rng= range(1,10)
sse = []
for k in k_rng:
    km= KMeans(n_clusters=k)
    km.fit(df[['age','income']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('sum of square error')
plt.plot(k_rng,sse)
```
